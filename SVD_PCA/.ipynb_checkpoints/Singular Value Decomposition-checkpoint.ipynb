```python
from warnings import filterwarnings 
filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats
from scipy.linalg import svd
#from numpy.linalg import svd

pd.set_option('display.width', 100)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python

df = pd.read_csv("../data/boston_house_prices.csv")
describe_df = df.describe()
descripe_std_df = describe_df.loc["std"].sort_values(ascending=False)
reorg_df = df[descripe_std_df.index]

print(reorg_df.head())
print("\n")

U, S, Vt = svd(reorg_df, full_matrices=False)
num_sv = np.arange(1, S.size+1)

cum_var_explained = [np.sum(S[0:n]) / np.sum(S) for n in num_sv]
print("cum_var_explained")
print(cum_var_explained)
print("\n")

# sdiag = np.diag(S)
sdiag = [np.sum(S[n]) for n in num_sv-1]
print("sdiag")
print(sdiag)
print("\n")

fig = plt.figure(figsize=(5.0,2.5))
ax = fig.add_subplot(111)
plt.plot(sdiag,
         color='#2171b5',
         label='variance')

plt.legend(loc='center right', scatterpoints=1, fontsize=8)
ax.set_xlabel('Number of singular values used')

plt.grid(alpha=0.5, zorder=1)
plt.tight_layout()
```
